# KEY FUNCTION 

```python


from selenium import webdriver
# didn't end up needing
# from selenium.webdriver.common.by import By
# from selenium.webdriver.support.ui import WebDriverWait # available since 2.4.0
# from selenium.webdriver.support import expected_conditions as EC # available since 2.26.0
# from selenium.webdriver.common.keys import Keys
import time

import os

chromedriver = "/home/douglas/Downloads/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)
driver.implicitly_wait(10) # seconds - letting page load before scraping, else timeout

def get_plantnet_urls(driver, plant, url):
    '''takes a url for a Pl@ntNet species, iterate through organ tabs and scrape all image links 
    Requires: driver: selenium driver, 
              plant: name of species,
              url for Pl@ntNet page'''
    driver.get(url)
    contents = []
    #replaced with implicitly_wait
#     try:
#         element = WebDriverWait(driver, 10).until(
#             EC.element_to_be_clickable(By.CLASS_NAME, "icon-tag"))
    
    for ele in driver.find_elements_by_class_name('icon-tag'):
        ele.click() #click through organ tabs
        time.sleep(1) #wait a second for images to load
        organ = ele.get_attribute('ng-src').split('/')[-1].split('.')[0] #get plant organ from url
        contents.append(organ)
        for pic in driver.find_elements_by_class_name('img-thumbnail'): #get links from each tab
            contents.append(pic.get_attribute('ng-src').replace('/s/', '/m/')) #/m/ holds the full size imgs
    with open('{}.txt'.format(plant), 'w') as file: # w vs wb - write vs write bytes(pics)
        for content in contents:
            file.write(str(content) + "\n") # create txt file with list of urls and organ as separator
#     finally:
    driver.quit() # close chrome
                
get_plantnet_urls(driver, 'pokeweed', 'https://identify.plantnet-project.org/species/namerica/Phytolacca%20americana%20L.')


```

# Adapted Imagenet scraper for new url lists 

```python
#Iterative scraper
import requests
import io
from PIL import Image
from requests.exceptions import ConnectionError


l = legumes + forbs # counting legumes as forbs
for species in l: # each species has a txt files with urls
    path = '/home/douglas/ds/metis/metisgh/Final project/Forbs/{}.txt'.format(species)
    plant = species
    # count images scraped and errors 
    n = 0
    se = 0
    ce = 0
    OSe = 0
    with open(path) as f:
        for line in f:
    #         if n > 5:
    #             break
            url = line.strip()
            if not url.startswith('https'): #  headings were mixed in with the urls to separate the urls
                organ = url # update organ label
                continue
            else:
                #print(url)
                try:
                    r = requests.get(url, data=None, headers={
                                    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2227.1 Safari/537.36'},
                                    timeout=2)
                    #print(r.connection, r.status_code)

                    i = Image.open(io.BytesIO(r.content)) # open as an array
                    print(i.size)
                    o_wd, o_ht = i.size[0], i.size[1]
                    # if square, resize to 224,224 for VGG
                    if o_ht == o_wd:
                        i = i.resize((224, 224), Image.ANTIALIAS)
                    # if rectangular, keep aspect ratio and scale smaller side to 256 
                    # will take random crops from this rectangle later as input to CNN
                    elif o_ht > o_wd: 
                        ratio = 256/o_wd
                        hsize = int(o_ht*ratio) 
                        wsize = int(o_wd*ratio)
                        i = i.resize((wsize, hsize), Image.ANTIALIAS)
                    else:
                        ratio = 256/o_ht
                        hsize = int(o_ht*ratio) 
                        wsize = int(o_wd*ratio)
                        i = i.resize((wsize, hsize), Image.ANTIALIAS)
                    try:
                        # image name has both plant and organ, so can separate further if useful
                        i.save('/home/douglas/ds/metis/metisgh/Final project/{}_{}{}.jpeg'.format(plant, organ, str(n)), quality=95)
                        n += 1
                        time.sleep(1) # polite scraping
                    except: # error with saving
                        
                        se += 1
                        print('se:', str(se))
                        continue
                except ConnectionError: # catch website error
                    ce += 1
                    print('ce:', ce)
                    continue
                except OSError: 
                    OSe += 1
                    print('OSe:', OSe)
                    continue

    print('n, se, ce, OSe')
    print(n, se, ce, OSe) # pl@ntnet images are all up, unlike some of links from imagenet that are outdated
```
